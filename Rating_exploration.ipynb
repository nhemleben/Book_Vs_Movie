```python
# Python
import pandas as pd

# Load the CSV file
df = pd.read_csv('adaptations_with_ratings.csv')

# Display the first few rows
df.head()
```

```python
# Sort by source_score (descending) and display the unique top 100 source values
top_sources = df.sort_values("source_score", ascending=False)[["source","source_score"]].drop_duplicates().head(50)
print(top_sources)
```

```python
# For each unique source, get its score and the adaptation with the highest rating
# (If all ratings are -1 for a source, it will pick the first adaptation.)

# Convert rating to numeric (in case it's not)
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

# Sort by source, then by rating descending
df_sorted = df.sort_values(["source", "rating"], ascending=[True, False])

# Drop duplicates to keep only the adaptation with the highest rating for each source
small_df = df_sorted.drop_duplicates(subset=["source"])

# Keep only relevant columns
small_df = small_df[["source", "source_score", "title", "url", "rating"]]

# Reset index for cleanliness
small_df = small_df.reset_index(drop=True)

# Display the smaller DataFrame
small_df.head()
```

```python
# Remove rows where source_score is -1 or the rating is -1 (indicating no data found)
small_df = small_df[small_df["source_score"] != -1].reset_index(drop=True)
small_df = small_df[small_df["rating"] != -1].reset_index(drop=True)

# Display the filtered DataFrame
small_df.head()
```

```python
better_adaptations = small_df[small_df["rating"] > small_df['source_score']].reset_index(drop=True)
better_adaptations.head()
```

```python
worse_adaptations = small_df[small_df["rating"] < small_df['source_score']].reset_index(drop=True)
worse_adaptations.head()
```

```python
equal_adaptations = small_df[small_df["rating"] == small_df['source_score']].reset_index(drop=True)
equal_adaptations.head()
```

```python
print("Total unique sources with valid scores:", len(small_df))
print(f"Number of better adaptations: {len(better_adaptations)}")
print(f"Number of worse adaptations: {len(worse_adaptations)}")
print(f"Number of equal adaptations: {len(equal_adaptations)}")
```
